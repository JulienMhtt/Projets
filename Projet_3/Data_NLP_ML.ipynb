## préparation de la colonne rdfs:comment.fr pour son utilisation dans le ML

```python
import nltk
nltk.download('popular')
```

```python
import spacy
import string
from nltk.tokenize.punkt import FreqDist
import re

nlp = spacy.load('fr_core_news_lg')

def clean(sentence):
    """
    This function 'clean' takes a single input 'sentence', a string, and performs several operations to 'clean' the text.

    These operations are:
    1. Removing punctuation using a regular expression.
    2. Removing accent
    3. Tokenizing the sentence into individual words.
    4. Removing English stopwords.
    5. Joining the remaining words back into a sentence.
    6. Lemmatizing the sentence using the Spacy English language model.

    The function returns the cleaned, lemmatized sentence as a single string.
    """
    # Retrait de la ponctuation
    sentence_no_punct = re.sub(r"[^\w\s]", "", sentence)
    
    # Retrait des accents
    # Définition du dictionnaire pour remplacer les caratères
    dict_replacement = {"à": "a", "â": "a", "ä": "a", "é": "e", "è": "e", "ê": "e", "ë": "e",
                    "î": "i", "ï": "i", "ô": "o", "ö": "o", "ù": "o", "û": "u", "ü": "u"}

    # Création d'une fonction qui permet de changer les caractères à accent
    # La fonction reçoit la lettre de la correspondance trouvée
    def remplacer(match):
        # La lettre correspondante au regex est extraite
        lettre = match.group(0)
        #Renvoi la valeur correspondante du dictionnaire si celle ci est présente
        return dict_replacement.get(lettre)

    sentence_no_punct_no_accent = re.sub(r"[àâäéèêëîïôöùûü]", remplacer, sentence_no_punct)

    # Transformation de la phrase en tokens
    sentence_no_punct_no_accent = nltk.word_tokenize(sentence_no_punct_no_accent.lower())

    # Liste des stopwords à retirer
    stop_words = set(nltk.corpus.stopwords.words("french"))

    # Récupération des tokens sans stopwords
    sentence_clean = []
    for words in sentence_no_punct_no_accent:
        if words not in stop_words:
            sentence_clean.append(words)

    # Transformer la liste en chaine de caratères
    sentence_clean_str = ' '.join(sentence_clean)

    # Utilisation de spacy
    sentence_clean_str_spacy = nlp(sentence_clean_str)

    sentence_clean_str_spacy_lemmatized = [token.lemma_ for token in sentence_clean_str_spacy]

    return ' '.join(sentence_clean_str_spacy_lemmatized)

dfs_verif["rdfs:comment.fr"] = dfs_verif["rdfs:comment.fr"].apply(clean)
```
